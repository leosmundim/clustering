## Padronização das Variáveis - ZScore 

- Base com variáveis em unidades de medidas diferentes, necessário aplicar a padronização para evitar distorções;

- Aplica-se o ZScore:
> Xi - Xmédia / Desv. Padrao

```python
scaler = StandardScaler()

#Selciona as variáveis
variaveis = dadosUso.iloc[:,2:]

#Padroniza dados para z-score
dadosUsoPad = pd.DataFrame(scaler.fit_transform(variaveis))

dadosUsoPad.columns = variaveis.columns

#Visualiza base
dadosUsoPad
```

```python
from mpl_toolkits.mplot3d import Axes3D

#Gráfico 3D
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(dadosUsoPad['Total_Credit_Cards'],
           dadosUsoPad['Avg_Credit_Limit'],
           dadosUsoPad['Total_visits_bank']
          )
ax.set_xlabel('#Cartões')
ax.set_ylabel('Limite Médio')
ax.set_zlabel('#Visitas')
ax.view_init(10, 60)
```

## Método Elbow

Verificando número ótimo de clusters

```python
from sklearn.cluster import KMeans

sum_sqr=[]
max_clusters = 10
K = range(1,max_clusters)

for k in K:
    model = KMeans(n_clusters=k)
    model.fit(dadosUsoPad)
    sum_sqr.append(model.inertia_)
```

```python
#Gráfico plotando Elbow
plt.figure(figsize=(16,8))
plt.plot(K, sum_sqr, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.axvline(x=3, c='r',ls='--')
plt.show()
```

```python
#Adicionando variável à base
cluster_K = KMeans(n_clusters=3).fit(dadosUsoPad)

dadosUsoPad['cluster_K'] = cluster_K.labels_
```

## Estatística Descritiva dos Clusters

```python
#Inserindo Clusters na base original
dadosUso['cluster_K'] = cluster_K.labels_

#Sumarizando e tirando a média por variável
dadosUso.groupby('cluster_K')['Total_Credit_Cards', 
                              'Avg_Credit_Limit',
                              'Total_visits_online',
                              'Total_visits_online',
                              'Total_calls_made'].mean()
```

## Análise de Variância de um Fator (ANOVA)

**Identidicar as Variáveis que mais ajudaram na formação dos clusters.**

- Mean Sq do Cluster_H: indica a variabilidade entre grupos
- Mean Sq dos Residuals: indica a variabilidade dentro dos grupos
- F Value: estatística de teste (Sum sq do Cluster_H / Sum Sq dos Resíduals)
- Pr(>F): p-valor da estatística
- p-valor < 0.05: rejeita-se H0 

>**H0:** A variável não é relevante para a separação dos clusters

>**H1:** A variável é relevante para a separação do Cluster

**Objetivo é comparar variabilidade entre os grupos e variabilidade dentro dos grupos** 

Variabilidade entre grupos tem que ser maior do que variabilidade dentro dos grupos, ou seja, os grupos devem ser diferentes entre si para formar clusters bem distintos.

**Estatística F:**
> Variabilidade entre grupos / variabilidade dentro do grupo

ou

> mean_sq (Cluster_H) / mean_sq (Residual)

**A variável mais discriminante dos grupos contém maior estatística F (e significativa: p-valor < 0.05)**

```python
# Referência: https://www.reneshbedre.com/blog/anova.html
from bioinfokit.analys import stat

res = stat()
print('ANOVA: Limite Médio')
res.anova_stat(df=dadosUsoPad, res_var='Avg_Credit_Limit', 
               anova_model=f'Avg_Credit_Limit ~ C(cluster_K)')
res.anova_summary
```

```python
print('ANOVA: Qtd Cartões')
res.anova_stat(df=dadosUsoPad, res_var='Total_Credit_Cards', 
               anova_model=f'Total_Credit_Cards ~ C(cluster_K)')
res.anova_summary
```

```python
print('ANOVA: Total de Visitas no Banco')
res.anova_stat(df=dadosUsoPad, res_var='Total_visits_bank', 
               anova_model=f'Total_visits_bank ~ C(cluster_K)')
res.anova_summary
```

```python
print('ANOVA: Total de Visitas Online')
res.anova_stat(df=dadosUsoPad, res_var='Total_visits_online', 
               anova_model=f'Total_visits_online ~ C(cluster_K)')
res.anova_summary
```

```python
print('ANOVA: Total de Ligações')
res.anova_stat(df=dadosUsoPad, res_var='Total_calls_made', 
               anova_model=f'Total_calls_made ~ C(cluster_K)')
res.anova_summary
```
