## K-Means

### Método de Elbow

Analisando a quantidade de clusters ótimo pelo método Elbow

```python
from sklearn.cluster import KMeans

variaveis = regionalVarejista[['atendimento', 'sortimento', 'organização']]

sum_sqr=[]
max_cluster = regionalVarejista.index.nunique()
K = range(1,max_cluster)

for k in K:
    model = KMeans(n_clusters=k)
    model.fit(variaveis)
    sum_sqr.append(model.inertia_)
    
```

```python
#Gráfico plotando Elbow
plt.figure(figsize=(16,8))
plt.plot(K, sum_sqr, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.axvline(x=3, c='r',ls='--')
plt.show()
```

## Cluster Hierarquico Aglomerativo

Analisando a quantidade de clusters ótimos pelo Método Hierárquico Aglomerativo

```python
#https://docs.scipy.org/doc/scipy/reference/cluster.hierarchy.html
import scipy.cluster.hierarchy as shc

#Medida de Distância sokalmichener
#https://docs.scipy.org/doc/scipy/reference/generated/scipy.spatial.distance.pdist.html#scipy.spatial.distance.pdist

## "complete": encadeamento completo (furthest neighbor ou complete linkage)
## "single": encadeamento único (nearest neighbor ou single linkage)
## "average": encadeamento médio (between groups ou average linkage)

clusters = shc.linkage(variaveis,
                      method='single',
                      metric='euclidean')
```

```python
#Plota Dendrograma
plt.figure()
shc.dendrogram(Z=clusters, 
               orientation='top', 
               distance_sort='descending', 
               count_sort='descending',
              )
plt.axhline(y=38, c='r',ls='--')
plt.show()
```

## Aplicando K-Means e Cluster Hierárquico

```python
#Aplicando Clusters K-Means

cluster_K = KMeans(n_clusters=3).fit(variaveis)

regionalVarejista['cluster_K'] = cluster_K.labels_
```

```python
#Aplicando Clusters Método Aglomeração Hierárquica
cluster_H = shc.cut_tree(clusters, n_clusters=3)

regionalVarejista['cluster_H'] = cluster_H
```

## Resultado Final com a Classificação dos Dois Métodos

```python
regionalVarejista.sort_values('regional')
```

## Análise de Variância de um Fator (ANOVA)

**Identidicar as Variáveis que mais ajudaram na formação dos clusters.**

- Mean Sq do Cluster_H: indica a variabilidade entre grupos
- Mean Sq dos Residuals: indica a variabilidade dentro dos grupos
- F Value: estatística de teste (Sum sq do Cluster_H / Sum Sq dos Resíduals)
- Pr(>F): p-valor da estatística
- p-valor < 0.05: rejeita-se H0 

>**H0:** A variável não é relevante para a separação dos clusters

>**H1:** A variável é relevante para a separação do Cluster

**Objetivo é comparar variabilidade entre os grupos e variabilidade dentro dos grupos** 

Variabilidade entre grupos tem que ser maior do que variabilidade dentro dos grupos, ou seja, os grupos devem ser diferentes entre si para formar clusters bem distintos.

**Estatística F:**
> Variabilidade entre grupos / variabilidade dentro do grupo

ou

> mean_sq (Cluster_H) / mean_sq (Residual)

**A variável mais discriminante dos grupos contém maior estatística F (e significativa: p-valor < 0.05)**

## ANOVA CLUSTER HIERÁRQUICO

```python
# Referência: https://www.reneshbedre.com/blog/anova.html
from bioinfokit.analys import stat

res = stat()
print('ANOVA: Atendimento')
res.anova_stat(df=regionalVarejista, res_var='atendimento', anova_model=f'atendimento ~ C(cluster_H)')
res.anova_summary
```

```python
print('ANOVA: Sortimento')
res.anova_stat(df=regionalVarejista, res_var='sortimento', anova_model=f'sortimento ~ C(cluster_H)')
res.anova_summary
```

```python
print('ANOVA: Organização')
res.anova_stat(df=regionalVarejista, res_var='organização', anova_model=f'organização ~ C(cluster_H)')
res.anova_summary
```

## Análise de Robustez

Aplicando o método hierárquico utilizando medida de distância de <i>manhattan</i> e encadeamento <i>complete</i>

```python
clusters_man = shc.linkage(variaveis,
                      method='complete',
                      metric='cityblock')
```

```python
#Plota Dendrograma
plt.figure()
shc.dendrogram(Z=clusters_man, 
               orientation='top', 
               distance_sort='descending', 
               count_sort='descending',
              )
plt.axhline(y=100, c='r',ls='--')
plt.show()

# O RESULTADO TAMBÉM APONTA PARA 3 CLUSTERS
```

```python
#Adicionando os clusters à base

regionalVarejista['cluster_H_man'] = shc.cut_tree(clusters_man, n_clusters=3)
regionalVarejista.sort_values('regional')
```
